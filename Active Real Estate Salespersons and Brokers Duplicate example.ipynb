Active Real Estate Salespersons and Brokers Duplicate example


```python
#Import library
import pandas as pd
```

```python
#Read the dataframe

df = pd.read_csv("Active_Real_Estate_Salespersons_and_Brokers (Edit 8.2).csv")

df.info()
df.head()
```

```python
#drop duplicates

df.drop_duplicates()
```

```python
#drop duplicates holding the Business Name as the leading column

df.drop_duplicates(subset=['Business Name'])
```

```python
#drop duplicates holding the Business Name & Business ZIP as the leading columns

df.drop_duplicates(subset=['Business Name', 'Business Zip'])
```

```python
#drop duplicates holding the Business Name & Business ZIP as the leading columns and keeping the last record

df.drop_duplicates(subset=['Business Name', 'Business Zip'], keep='last')
```
